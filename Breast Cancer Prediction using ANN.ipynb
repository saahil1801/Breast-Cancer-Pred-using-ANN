```python
import numpy as np #Importing necessary libraries
```

```python
import pandas as pd  #Importing necessary libraries
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
from sklearn.model_selection import train_test_split 
```

```python
# Load the Breast Cancer dataset from the sklearn.datasets module
df1=sklearn.datasets.load_breast_cancer()
```

```python
# Create a DataFrame 'df' using the data and feature names from the Breast Cancer dataset
df = pd.DataFrame(df1.data, columns=df1.feature_names)
```

```python
df
```

```python
# Add a 'label' column to the DataFrame 'df' containing the target values from the Breast Cancer dataset
df['label'] = df1.target
```

```python
df
```

```python
#Displays details about its columns, data types, non-null counts, and memory usage.
df.info()
```

```python
# Separating features (x) and target labels (y)
x = df.drop(columns='label', axis=1)  # Extracting features by dropping the 'label' column
y = df['label']                      # Extracting target labels from the 'label' column
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# Feature Scaling using StandardScaler
from sklearn.preprocessing import StandardScaler

# Creating a StandardScaler object
scaler = StandardScaler()

# Fitting and transforming the training data
x_train = scaler.fit_transform(x_train)

# Transforming the test data using the same scaler
x_test = scaler.transform(x_test)
```

```python
# Importing TensorFlow and create a Sequential model for the ANN
import tensorflow as tf
```

```python
# Adding a Flatten layer to the model with an input shape of (30,)
ann=tf.keras.models.Sequential()
```

```python
# Adding a Flatten layer to the model with an input shape of (30,)
ann.add(tf.keras.layers.Flatten(input_shape=(30,)))
```

```python
# Adding a Dense layer with 20 units and ReLU activation to the ANN
ann.add(tf.keras.layers.Dense(units=20, activation='relu'))
```

```python
# Adding a Dense layer with 2 units and sigmoid activation for binary classification
ann.add(tf.keras.layers.Dense(units=2, activation='sigmoid'))
```

```python
# Compiling the artificial neural network (ANN) model
ann.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#Training
ann.fit(x_train,y_train,epochs=15)
```

```python
ann.evaluate(x_test,y_test)
```
